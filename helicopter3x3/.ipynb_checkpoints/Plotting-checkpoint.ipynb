# Lewis's Additions

I want to see how close each policy (extracted from the Q function learnt by the DQN) is to uniform across each region.

```python
policy_reg = []
policy_no_reg = []

for i in range(10):
    policy_reg.append(pd.read_csv("./policy_dists/policy_reg_{}.csv".format(i), index_col=0, names=list(range(10))))
    policy_no_reg.append(pd.read_csv("./policy_dists/policy_no_reg_{}.csv".format(i), index_col=0, names=list(range(10))))

fig2 = plt.figure()
fig2.suptitle('Example Policy Distributions')
plt.subplot(1, 2, 1)
s.heatmap(policy_reg[0], vmin=0, vmax=1, yticklabels=False, cmap="YlGnBu")
plt.subplot(1, 2, 2)
s.heatmap(policy_no_reg[0], vmin=0, vmax=1, yticklabels=False, cmap="YlGnBu")

concat_reg = pd.concat(policy_reg)
concat_no_reg = pd.concat(policy_no_reg)

avg_reg = concat_reg.groupby(concat_reg.index).mean()
avg_no_reg = concat_no_reg.groupby(concat_no_reg.index).mean()

fig2 = plt.figure()
fig2.suptitle('Averaged Policy Distributions')
a = plt.subplot(1, 2, 1)
a.set_xlabel('Actions')
a.set_ylabel('Regions')
s.heatmap(avg_reg, vmin=0, vmax=1, yticklabels=False, cmap="YlGnBu")
plt.subplot(1, 2, 2)
s.heatmap(avg_no_reg, vmin=0, vmax=1, yticklabels=False, cmap="YlGnBu")

reg_ent = sum([df.apply(sp.stats.entropy, axis=1).mean() for df in policy_reg]) / 10
no_reg_ent = sum([df.apply(sp.stats.entropy, axis=1).mean() for df in policy_no_reg]) / 10                         

print(reg_ent)
print(no_reg_ent)        
```
